```python
ang = open("0328-ang-1.txt")
jerk = open("0328-jerk-1.txt")
mul = open("0328-mul-1.txt")
list_ang  = ang.readlines()
list_jerk = jerk.readlines()
list_mul = mul.readlines()
```

```python
import re
ang_data=[]
jerk_data=[ ]
mul_data =[]
for list in list_ang:
    ang_data.append(re.findall("-?\d+\.?\d*",list))
for list in list_jerk:
    jerk_data.append(re.findall("-?\d+\.?\d*",list))
for list in list_mul:
    mul_data.append(re.findall("-?\d+\.?\d*",list))
```

```python
import pandas as pd

df_ang = pd.DataFrame(ang_data)
df_jerk = pd.DataFrame(jerk_data)
df_mul = pd.DataFrame(mul_data)
```

```python
import numpy as np

x_ang = np.array(range(0,len(df_ang)))[400:3000]
y_ang_ax = pd.to_numeric(df_ang[0])[400:3000]
y_ang_ay = pd.to_numeric(df_ang[1])[400:3000]
y_ang_az = pd.to_numeric(df_ang[2])[400:3000]

x_jerk = np.array(range(0,len(df_jerk)))[500:3100]
y_jerk_ax = pd.to_numeric(df_jerk[0])[500:3100]
y_jerk_ay = pd.to_numeric(df_jerk[1])[500:3100]
y_jerk_az = pd.to_numeric(df_jerk[2])[500:3100]

x_mul = np.array(range(0,len(df_mul)))[800:3400]
y_mul_ax = pd.to_numeric(df_mul[0])[800:3400]
y_mul_ay = pd.to_numeric(df_mul[1])[800:3400]
y_mul_az = pd.to_numeric(df_mul[2])[800:3400]
```

```python
print(np.std(y_ang_az))
print(np.std(y_jerk_az))
print(np.std(y_mul_az))
```

```python
import matplotlib.pyplot as plt
fig_ang=plt.subplot(3,1,1)
fig_ang.plot(x_ang,y_ang_ax)
fig_ang.plot(x_ang,y_ang_ay)
fig_ang.plot(x_ang,y_ang_az)

fig_jerk=plt.subplot(3,1,2)
fig_jerk.plot(x_jerk,y_jerk_ax)
fig_jerk.plot(x_jerk,y_jerk_ay)
fig_jerk.plot(x_jerk,y_jerk_az)

fig_mul=plt.subplot(3,1,3)
fig_mul.plot(x_mul,y_mul_ax)
fig_mul.plot(x_mul,y_mul_ay)
fig_mul.plot(x_mul,y_mul_az)
```
